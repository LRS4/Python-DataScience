# Investment Modelling

First model is standard exponential growth, second model is regular contributions to exponential growth.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# variables
rate = 1.07
years_saving = 40
initial_savings = 20000

# charting
model = pd.DataFrame({'t': range(years_saving)})
model['simple_exp'] = [initial_savings*rate**year for year in model.t]

plt.figure()
plt.plot(model.t, model.simple_exp)
plt.title('Exponential Growth of Initial Investment')
plt.xlabel('time (years)')
plt.ylabel('value (£)')

# text assessment
final_amount = int(initial_savings*(rate**years_saving))
rate_1 = round(((rate-1)*100),0)
print("After " + str(years_saving) + " years the initial investment of £" + str(initial_savings) + " has grown by " + str(rate_1) + " percent each year to £" + str(final_amount))
```

```python
# equation for regular contributions to an exponentially growing investment
# value = P*ry + C*(ry - 1)/(r - 1)

# variable
yearly_contribution = 650*12

# charting
model['yearly_invest'] = model['simple_exp'] + [yearly_contribution*(rate**year - 1)/(rate-1) for year in model.t]
plt.plot(model.t, model.yearly_invest)
plt.title('Exponential Growth of Yearly £' + str(yearly_contribution) + ' Contribution')
plt.legend(['initial savings', 'yearly contribution'])

# text assessment
yearly_final = int(final_amount + yearly_contribution*(rate**years_saving -1)/(rate - 1))
print("An initial investment of £" + str(initial_savings) + " alongside a yearly contribution of £" + str(yearly_contribution) + " results in £" + str(yearly_final) + " after " + str(years_saving) + " years")
```
